```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('Car_sales.csv')
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
missing_count = df[["Model","Sales_in_thousands"]].isnull().sum()
print(missing_count)
```

```python
df = df.dropna(subset=["Model","Sales_in_thousands"])
df
```

```python
df = df.groupby("Model")["Sales_in_thousands"].sum().reset_index()
df
```

```python
df = df.sort_values(
    by="Sales_in_thousands",
    ascending=False
)
```

```python
df = df.head(10)
df
```

```python
plt.figure(figsize=(10,6))
plt.bar(df["Model"],df["Sales_in_thousands"])
plt.xlabel("Models")
plt.ylabel("Sales_in_thousands")
plt.title("Top 10 Models by Sales")
plt.xticks(rotation=45)
plt.show
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    df["Sales_in_thousands"],
    labels=df["Model"],
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Sales Distribution by Car Model")
plt.show()
```

```python
fig = px.bar(
    df,
    x="Model",
    y="Sales_in_thousands",
    title="Car Sales by Model",
    labels={"TOTAL SALES"}
)
fig.show()
```

```python
fig = px.pie(
    df,
    names="Model",
    values="Sales_in_thousands",
    title="Sales Distribution by Car Model",
    labels={"Sales_in_thousands": "Total Sales (in thousands)"}
)
fig.show()
```
